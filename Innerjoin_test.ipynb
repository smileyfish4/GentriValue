```python
import pandas as pd 
```

```python
#original excel sheet

og_df = pd.DataFrame(pd.read_csv("KYS.csv"))
og_df.rename(columns = {'Geo_ID' : 'GeoID'}, inplace = True)
og_df.rename(columns = {'change of median rent' : 'percent_change_md_rent00_16'}, inplace = True)
column_reorder = ['GeoID','NYC Neighborhood Tabulation Area name', 'percent_change_md_rent00_16', 'College00_16', 'mdfami00_16']
og_df.head()
og_df.iloc[:20]
og_df.dropna()
og_df1 = og_df[column_reorder]
og_df1.dtypes
 
```

```python
#per capita income

per_capita_income_00 = pd.DataFrame(pd.read_csv("per_capita_income_00.csv"))
per_capita_income_00_1 = per_capita_income_00.iloc[1:] 
per_capita_income_00_2 = per_capita_income_00_1.dropna()
per_capita_income_00_3 = per_capita_income_00_2[['GeoID', 'Per Capita Income']]
per_capita_income_00_4 = per_capita_income_00_3.astype({'GeoID': 'int', 'Per Capita Income' : 'float'})


```

```python
per_capita_income_20_2 = pd.DataFrame(pd.read_csv('per_capita_income_20_2.csv'))
per_capita_income_20_2.rename(columns = {'apci': 'Per Capita Income 2020'}, inplace = True)
per_capita_income_20_3= per_capita_income_20_2.dropna()
per_capita_income_20_4 = per_capita_income_20_3.iloc[1:]
per_capita_income_20_5 = per_capita_income_20_4[['GeoID', 'Per Capita Income 2020']]
per_capita_income_20_5.dtypes
```

```python
per_capita_income_merge = pd.merge(per_capita_income_20_5, per_capita_income_00_4,on = 'GeoID', how = 'inner')
test_merge = pd.merge(og_df1, per_capita_income_merge, on = 'GeoID', how = 'inner')
test_merge
```

```python
def roc(per_capita_income_merge, per_capita_income_00_4, per_capita_income_20_5, roc_per_capita_income):
    per_capita_income_merge['roc_pci'] = (per_capita_income_merge['per_capita_income_20_5'] - per_capita_income_merge['per_capita_income_00_4']) / per_capita_income_merge['per_capita_income_00_4']
roc(per_capita_income_merge)
```

```python
black_pop_00 = pd.DataFrame(pd.read_csv('black_pop_00.csv'))
black_pop_00_1 = black_pop_00.iloc[1:]
black_pop_00_2 = black_pop_00_1.dropna()
black_pop_00_3 = black_pop_00_2[['GeoID', 'Percent Black or African American Population']]
black_pop_00_4 = black_pop_00_3.rename(columns = {'Percent Black or African American Population' : 'Percent Black or African American Population 2000'})
black_pop_00_4
```

```python
black_pop_20 = pd.DataFrame(pd.read_csv('black_pop_20.csv'))
black_pop_20_1 = black_pop_20.iloc[1:]
black_pop_20_2 = black_pop_20_1.dropna()
black_pop_20_3 = black_pop_20_2[['GeoID', 'Percent Black or African American Population']]
black_pop_20_4 = black_pop_20_3.rename(columns = {'Percent Black or African American Population' : 'Percent Black or African American Population 2020'})
black_pop_20_4
```

```python
black_pop_merge = pd.merge(black_pop_00_4, black_pop_20_4, on = "GeoID", how = 'inner')
black_pop_merge_dtype = black_pop_merge.astype({'GeoID' : 'int', 'Percent Black or African American Population 2000': 'float', 'Percent Black or African American Population 2020' : 'float'})
black_pop_merge_dtype.dtypes
test_merge_2 = pd.merge(test_merge, black_pop_merge_dtype, on = "GeoID", how = 'inner')
test_merge_2
```
